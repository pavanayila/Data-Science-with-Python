# Pie Chart

```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

# Variable name called slices
slices = [60, 40]

plt.pie(slices)

plt.title('My Awesome Pie Chart')

plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

# Variable name called slices
slices = [60, 40]
labels = ['sixty', 'forty']

plt.pie(slices, labels = labels)

plt.title('My Awesome Pie Chart')

plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

# Variable name called slices
slices = [60, 40]
labels = ['sixty', 'forty']

# To change the edge color the wedge properties takes the dictionary of values
plt.pie(slices, labels = labels, wedgeprops = {'edgecolor' : 'black'})

plt.title('My Awesome Pie Chart')

plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

# Variable name called slices
slices = [60, 40, 30, 20]
labels = ['sixty', 'forty', 'extra1', 'extra2']
colors = ['blue', 'red', 'yellow', 'green']

# To change the edge color the wedge properties takes the dictionary of values
plt.pie(slices, labels = labels, colors = colors, wedgeprops = {'edgecolor' : 'black'})

plt.title('My Awesome Pie Chart')

plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')

slices = [59219, 55466, 47544, 36443, 35917]

labels = ['Javascript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode = [0, 0, 0, 0.1, 0]

# auto push percent displays percentage on the pie chart
plt.pie(slices, labels = labels, explode = explode, shadow = True, startangle = 90, autopct = '%1.1f%%', wedgeprops = {'edgecolor' : 'black'})

plt.title('My Awesome Pie Chart')

plt.tight_layout()

plt.show()
```
